# What is the Doses-per-Session and Doses-per-Site trend over days?

```python
# What is the Doses-per-Session and Doses-per-Site trend over days?
covid_vaccine_df['Doses_per_Session'] = covid_vaccine_df['Total Doses Administered']/ covid_vaccine_df['Sessions']
covid_vaccine_df['Doses_per_Site'] = covid_vaccine_df['Total Doses Administered']/ covid_vaccine_df['Sites']
trend = covid_vaccine_df[['Updated On' , 'Doses_per_Session','Doses_per_Site']]
print(trend)
```
